```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Dimensiones DAMA
labels = [
    "Gobernanza", "Calidad", "Arquitectura",
    "Seguridad", "Integración", "Operaciones", "Analítica"
]

# Niveles (1 a 5)
empresa_x = [3, 2, 4, 3, 2, 3, 4]
nivel_ideal = [5, 5, 5, 5, 5, 5, 5]

# Preparación del gráfico
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
empresa_x += empresa_x[:1]
nivel_ideal += nivel_ideal[:1]
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, empresa_x, label="Empresa X", linewidth=2)
ax.fill(angles, empresa_x, alpha=0.25)

ax.plot(angles, nivel_ideal, label="Nivel Ideal", linestyle='dashed', color='grey')
ax.fill(angles, nivel_ideal, alpha=0.05, color='grey')

ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_ylim(0, 5)
ax.set_title("Evaluación de Madurez en Gestión de Datos (DAMA)", size=14)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```
